# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data

```python
dataset=pd.read_csv("data.csv")
```

```python
dataset.head()
```

```python
dataset
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset[dataset['DEATH_EVENT'] == 1]['age'], dataset[dataset['DEATH_EVENT'] == 1]['platelets'], c='green', label='Survive', alpha=0.5)
plt.scatter(dataset[dataset['DEATH_EVENT'] == 0]['age'], dataset[dataset['DEATH_EVENT'] == 0]['platelets'], c='red', label='Did Not Survive', alpha=0.5)
plt.xlabel('Age')
plt.ylabel('platelets')
plt.legend()
plt.title('Scatter Plot of Age vs platelets for Survived vs Did Not Survive')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', hue='DEATH_EVENT', data=dataset, palette='Set2')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Survival by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
X=dataset.drop("DEATH_EVENT",1)
y=dataset["DEATH_EVENT"]
```

```python
X
```

```python
X.describe()
```

# Removing Null values

```python
rows_with_null=X.isnull().any(axis=1)
dataset_cleaned=X.dropna()
print(dataset_cleaned.head())
```

# Splitting the data
**Training and Testing data**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y
```

# Applying classifier
**Decision Tree Classifier**

```python
my_model=DecisionTreeClassifier()
my_model.fit(X_train,y_train)
y_pred=my_model.predict(X_test)
```

**Mean Absolute Error**

```python
mae=mean_absolute_error(y_pred,y_test)
print("Mean absolute error is:- ",mae)
```

**Accuracy**

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.3f}\n')
```

**Confusion Matrix**

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(conf_matrix)
```

**Classification Report**

```python
class_report = classification_report(y_test, y_pred)
print("Classification Report:")
print(class_report)
```
